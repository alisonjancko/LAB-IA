```python
def  featureNormalize(X):
    X_norm = X.copy()
    mu = np.zeros(X.shape[1])
    sigma = np.zeros(X.shape[1])

    mu = np.mean(X, axis = 0)
    sigma = np.std(X, axis = 0)
    X_norm = (X - mu) / sigma
    
    return X_norm, mu, sigma
```

```python
# llama featureNormalize con los datos cargados
X_norm, mu, sigma = featureNormalize(X)

print(X)
print('Media calculada:', mu)
print('Desviacin estandar calculada:', sigma)
print(X_norm)
```

```python
def computeCostMulti(X, y, theta):
    # Inicializa algunos valores utiles
    m = y.shape[0] # numero de ejemplos de entrenamiento
    
    J = 0
    
    h = np.dot(X, theta)
    
    J = (1/(2 * m)) * np.sum(np.square(np.dot(X, theta) - y))
    
    return J
```

```python
def gradientDescentMulti(X, y, theta, alpha, num_iters):
    
    # Inicializa algunos valores 
    m = y.shape[0] # numero de ejemplos de entrenamiento
    
    # realiza una copia de theta, el cual ser acutalizada por el descenso por el gradiente
    theta = theta.copy()
    
    J_history = []
    
    for i in range(num_iters):
        theta = theta - (alpha / m) * (np.dot(X, theta) - y).dot(X)
        J_history.append(computeCostMulti(X, y, theta))
    
    return theta, J_history
```

```python
# Elegir algun valor para alpha (probar varias alternativas)
alpha = 0.01
num_iters = 5000

# inicializa theta y ejecuta el descenso por el gradiente
theta = np.zeros(9)
theta, J_history = gradientDescentMulti(X, y, theta, alpha, num_iters)

# Grafica la convergencia del costo
pyplot.plot(np.arange(len(J_history)), J_history, lw=2)
pyplot.xlabel('Numero de iteraciones')
pyplot.ylabel('Costo J')

# Muestra los resultados del descenso por el gradiente
print('theta calculado por el descenso por el gradiente: {:s}'.format(str(theta)))

# Estimar el precio para un automovil Toyota modelo GTX 2011 con caja Manual,
#Kilometraje 45128, con tipo de combustible a gasolina, con un Impuesto de 125,con mpg 50.4, con un tamao de motor 1.3 

#4510
X_array = [1,6,2019,1,1791,1,145,57.7,1.0]
#11995

X_array[1:9] = (X_array[1:9] - mu) / sigma
price = np.dot(X_array, theta)   # Se debe cambiar esto

print('El precio predecido un automovil Toyota modelo Auris 2019 con caja Manual, Kilometraje 1791, con tipo de combustible a gasolina, con un Impuesto de 145,con mpg 57.7, con un tamao de motor 1.0      (usando el descenso por el gradiente): ${:.0f}'.format(price))
```

```python
#Ecuacion de la Normal
# Cargar datos
data = np.genfromtxt('toyota.csv', delimiter=';' )
X = data[:, :8]
y = data[:, 8]
m = y.size
X = np.concatenate([np.ones((m, 1)), X], axis=1)
```

```python
def normalEqn(X, y):
  
    theta = np.zeros(X.shape[1])
    
    theta = np.dot(np.dot(np.linalg.inv(np.dot(X.T,X)),X.T),y)
    
    return theta
```

```python
# Calcula los parametros con la ecuacin de la normal
theta = normalEqn(X, y);

# Muestra los resultados optenidos a partir de la aplicacin de la ecuacin de la normal
print('Theta calculado a partir de la ecuacin de la normal: {:s}'.format(str(theta)));

# Estimar el precio para una casa de superficie de 1650 sq-ft y tres dormitorios
#6000
X_array = [1,7,2017,2,11450,2,0,74.3,1.8]
#19461
price = np.dot(X_array, theta) 


print('El precio predecido un automovil Toyota modelo Yaris 2017 con caja Automatica, Kilometraje 11450, con tipo de combustible a diesel, con un Impuesto de 0,con mpg 74.3, con un tamao de motor 1.8      (usando el descenso por el gradiente): ${:.0f}'.format(price))
```
